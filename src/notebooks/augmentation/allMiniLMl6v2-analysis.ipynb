```python
import pandas as pd
```

```python
augmented_df = pd.read_csv("../../../synonym_replacement_results.csv")
```

```python
# Get the lowest 20 similarity scores
lowest_scores = augmented_df['Similarity Score'].nsmallest(20)

for i in range(20):
    print(str(augmented_df.iloc[lowest_scores.index[i]]['Claim']))
    print(str(augmented_df.iloc[lowest_scores.index[i]]['Original Evidence']))
    print(str(augmented_df.iloc[lowest_scores.index[i]]['Evidence']))
    print("\n")

```
